```python
import pandas as pd
data1 = pd.read_csv('/home/kali/Downloads/lego-analysis-master/datasets/lego_sets.csv')
data2 = pd.read_csv('/home/kali/Downloads/lego-analysis-master/datasets/parent_themes.csv')
data3 = pd.read_csv('/home/kali/Downloads/lego-analysis-master/datasets/sets.csv')
data4 = pd.read_csv('/home/kali/Downloads/lego-analysis-master/datasets/themes.csv')
merged = data1.merge(data2,left_on='parent_theme',right_on='name')
merged.drop(columns='name_y',inplace=True)
merged = merged.merge(data3,left_on = 'name_x',right_on ='name')
merged.drop(columns='name',inplace=True)
merged = merged.merge(data4,left_on = 'id',right_on ='parent_id')
merged.drop(columns='parent_id',inplace=True)
merged.head(1)
```
